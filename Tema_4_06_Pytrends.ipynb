## Buscamos algunas tendencias en España

### Pytrends

```python
pytrends = TrendReq(hl='ES', 
                    tz =360)  
pytrends
```

```python
kw_list = ["bocadillo", "hamburguesa", "pizza"] 
kw_list
```

Alguna información útil:

   * kw_list: una lista de palabras clave para obtener datos 
     (límite de 5 palabras clave a la vez)
     
   * geo: la ubicación geográfica.
   
   * timeframe: el marco de tiempo de interés.

```python
# Construimos el payload
pytrends.build_payload(kw_list, 
                       timeframe='2015-01-01 2021-01-01', 
                       geo='ES')
```

### DataFrame (podríamos enviar datos a un excel / csv

```python
# Interes A lo largo del tiempo
# Almacenando la información en un DataFrame 
df = pytrends.interest_over_time()
```

### Información relevante del DataFrame

```python
df.tail()
```

### Algunas gráficas

```python
# plot all three trends in same chart
plt.figure(figsize=(12,6))
plt.plot(df.index,df.bocadillo,'k*')
plt.plot(df.index,df.hamburguesa,'r*')
plt.plot(df.index,df.pizza,'b*')
plt.legend(['bocadillo','hamburguesa','pizza'])
plt.show()
```

```python
# build list of keywords
kw_list = ["bocadillo", "pizza", "hamburguesa", "kebap"] 

# build the payload
pytrends.build_payload(kw_list, 
                       cat=None, 
                       timeframe='2021-01-01 2021-02-01', 
                       geo='ES') 

# get the interest over time
df = pytrends.interest_over_time()

# ploteo tendencias de comida over time
plt.figure(figsize=(12,6))
plt.plot(df.index,df.bocadillo,'k*')
plt.plot(df.index,df.pizza,'r*')
plt.plot(df.index,df.hamburguesa,'b*')
plt.plot(df.index,df.kebap,'g*')
plt.legend(['bocadillo','pizza','hamburguesa','kebap'])
plt.show()
```

### related topics

```python
# build the payload
pytrends.build_payload(kw_list=['pizza'], 
                       timeframe='2016-01-01 2021-02-01', 
                       geo='ES') 

# get related topics
df_rt = pytrends.related_topics()
df_rt
```

### related queries

```python
# build the payload
pytrends.build_payload(kw_list=['pizza'], 
                       timeframe='2021-01-01 2021-02-01', 
                       geo='ES') 

# get related queries
df_rq = pytrends.related_queries()

# display df_rq
print(df_rq)
```

**INTERÉS EN UNA ÉPOCA CONCRETA**

```python
# build list of keywords
kw_list = ["Python", "R", "Matlab", "Julia"] 

# build the payload
pytrends.build_payload(kw_list, 
                       cat=None, 
                       timeframe='2019-01-01 2021-01-01', 
                       geo='US') 

# get the interest over time
df = pytrends.interest_over_time()

# ploteo tendencia en búsquedas por lenguaje de programación over time
plt.figure(figsize=(16,6))
plt.plot(df.index,df.Python,'k*')
plt.plot(df.index,df.R,'r*')
plt.plot(df.index,df.Matlab,'b*')
plt.plot(df.index,df.Julia,'g*')
plt.legend(['Python','R','Matlab','Julia'])
plt.show()
```

Si bien el interés a lo largo del tiempo es interesante, 

es posible que desee utilizar las tendencias de Google para encontrar:

   * temas, 
   * consultas 
   * o sugerencias relacionados. 
   
Para hacerlo, puede utilizar los métodos:

   * related_topics (), 
   * related_queries ()  
   * y Followers ().

```python
# build the payload
pytrends.build_payload(kw_list=['Python'], 
                       timeframe='2018-01-01 2021-01-01', 
                       geo='ES') 

# get related topics
df_rt = pytrends.related_topics()
print(df_rt)
```

```python
# top 5 rising results para la palabra clave Python
print(df_rt['Python']['rising'].head(5))
```

```python
# top 5 top results para la palabra clave Python
print(df_rt['Python']['top'].head(5))
```

```python
# build the payload
pytrends.build_payload(kw_list=['Python'], 
                       timeframe='2019-01-01 2021-01-01', 
                       geo='US') 

# get related queries
df_rq = pytrends.related_queries()

# display df_rq
print(df_rq)
```

```python
# Finalmente, podemos obtener palabras clave de búsqueda sugeridas 
# como se muestra a continuación:
```

```python
# get suggestions
df_s = pytrends.suggestions('Python')

# display df_rq
print(np.array(df_s))
```

# Y otro ejemplo más

```python
# Solo necesitamos pues ejecutar esto una vez
# el resto de requests usarán la misma sesión
pytrend = TrendReq()

# payload y API tokens. 
# solo necesario para interest_over_time(), interest_by_region() & related_queries()
pytrend.build_payload(kw_list=['pizza', 'kebab'])
```

```python
# Interest Over Time
interest_over_time_df = pytrend.interest_over_time()
print(interest_over_time_df.head())
```

```python
# Interest by Region
interest_by_region_df = pytrend.interest_by_region()
print(interest_by_region_df.head())
```

```python
# Related Queries, 
# returns a dictionary of dataframes
related_queries_dict = pytrend.related_queries()
print(related_queries_dict)
```

```python
# Get Google Hot Trends data
trending_searches_df = pytrend.trending_searches()
print(trending_searches_df.head())
```

```python
# Get Google Hot Trends data
today_searches_df = pytrend.today_searches()
print(today_searches_df.head())
```

```python
# Get Google Top Charts
top_charts_df = pytrend.top_charts(2018, 
                                   hl='en-US', 
                                   tz=300, 
                                   geo='GLOBAL')
print(top_charts_df.head())
```

```python
# Get Google Keyword Suggestions
suggestions_dict = pytrend.suggestions(keyword='pizza')
print(np.array(suggestions_dict))
```
