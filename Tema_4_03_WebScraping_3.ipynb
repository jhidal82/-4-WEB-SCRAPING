```python
import requests
from bs4 import BeautifulSoup
```

```python
page = requests.get('https://parascrapear.com/')
print(page)
print('\n')
soup = BeautifulSoup(page.text, 'html.parser')
print(soup)
```

```python
blockquote_items = soup.find_all('blockquote')
# print(blockquote_items)

for blockquote in blockquote_items:
    autor = blockquote.find(class_='author').text
    categoria = blockquote.find(class_='cat').text
    frase = blockquote.find('q').text
    print([autor, categoria, frase])
    print('\n')
```

# Trabajar la información en formato Dataframe

```python
import pandas as pd
```

```python
df = pd.DataFrame([], columns=["autor", "categoria", "frase"])
df.head()
```

```python
blockquote_items = soup.find_all('blockquote')
# print(blockquote_items)

for blockquote in blockquote_items:
    nueva_fila = {"autor": blockquote.find(class_='author').text,
                  "categoria": blockquote.find(class_='cat').text,
                  "frase": blockquote.find('q').text}
    df = df.append(nueva_fila, ignore_index = True)
df
```
